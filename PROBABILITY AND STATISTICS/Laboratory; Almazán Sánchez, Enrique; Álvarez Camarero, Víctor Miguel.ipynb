```python
#Creating the modules/functions need it to perform this exercise section.
def seedLCG(initVal):
    global rand
    rand = initVal

def lcg(a, b, m):
    global rand
    rand = ((a*rand + b) % m)
    return rand, m

def two_list(l):
    x_list = []
    y_list = []
    for element in l:
        x_list.append(element)
        y_list.append(element)
    x_list.pop(-1)
    y_list.pop(0)
    return x_list, y_list
```

**Exercise 7. Medical knee implant**

**A factory produces a piece of a medical knee implant. The length of each fabricated piece must be under control and the variance as low as possible. It is known that if the variance of the piece is less than 0.75 cm2 the process is under control. Knee piece implant length errors in the manufacturing process are known to be Gaussian distributed. A sample of the length of 100 hundred medical pieces is saved in the file implant.txt. Import the file in Matlab and compute a 99% confidence interval for the variance. Is the process under control?**


```python
import numpy
from scipy.stats import norm, kurtosis
from scipy.stats.distributions import chi2
```

```python
def list_data(f):
    data_list = open(f, mode='r')
    lines = [line.split() for line in data_list]
    total=list()
    for line in lines: 
        total+=line
    values = list(map(float, total))
    return values

def myUnbiasedVariance(data):
    # finds the 'unbiased' variance of a given set of numbers (divides by N-1) 
    xbar = numpy.average(data)
    return sum([(x - xbar)**2 for x in data])/(len(data)-1)
```

```python
values = list_data("data\exercise_7\implant.txt")

#Computing the mean
n = len(values)
mean = round(numpy.average(values), 3)
print("---MEAN---")
print("Mean of the sample --> ",  mean)

#Computing the standard deviation
std_deviation = round(numpy.std(values), 3)
print("\n---STANDARD DEVIATION---")
print("Standart deviation of the sample --> ",  std_deviation)

#Computing the variance
print("\n---VARIANCE---")
variance = round(numpy.var(values), 3)
print("Variance of the sample --> ",  variance)

#Computing the unbiased variance
unbiased_variance = round(myUnbiasedVariance(values),3)
print("\n---UNBIASED VARIANCE---")
print("Unbiased variance of the sample --> ",  unbiased_variance)

#Computing the critical values
critical_value1 = round(chi2.ppf(0.995, 99), 3)
critical_value2 = round(chi2.ppf(0.005, 99), 3)
print("\n---CRITICAL VALUES---")
print("Critical values --> ", critical_value1, " | ", critical_value2)

#Computing the confidence interval
first_number = round((((n - 1)* unbiased_variance)/critical_value1), 3)
second_number = round((((n - 1)* unbiased_variance)/critical_value2), 3)
print("\n---CONFIDENCE INTERVAL---")
print("The interval of the variance is: [" + str(first_number) + ", " + str(second_number) + "]")
```

```python
rad = cv2.imread('data\\exercise_11\\noisy_radiography.png', 0)
row, col = rad.shape
kernel = np.zeros([row, col])
```

(a) Implement one of the above filters, except the median. Choose a rule for the points on the border. What is the effect of the filter? Is the noise reduced?

```python
# MEAN FILTER
corner_mask = np.ones([2, 2], dtype = int)
corner_mask = corner_mask / 4

row_mask = np.ones([2, 3], dtype = int)
row_mask = row_mask / 6

col_mask = np.ones([3, 2], dtype = int)
col_mask = col_mask / 6

mask = np.ones([3, 3], dtype = int) 
mask = mask / 9

for i in range(row): 
    for j in range(col): 
        if i == 0 and j == 0:
            temp = rad[i, j]*corner_mask[0, 0]+rad[i+1, j]*corner_mask[1, 0]+rad[i, j+1]*corner_mask[0, 1]+rad[i+1, j+1]*corner_mask[1, 1]
            kernel[i, j] = temp
        elif i == 0 and j == col - 1:
            temp = rad[i, j-1]*corner_mask[0, 0]+rad[i+1, j-1]*corner_mask[1, 0]+rad[i, j]*corner_mask[0, 1]+rad[i+1, j]*corner_mask[1, 1]
            kernel[i, j] = temp
        elif i == row - 1 and j == 0:
            temp = rad[i-1, j]*corner_mask[0, 0]+rad[i, j]*corner_mask[1, 0]+rad[i-1, j+1]*corner_mask[0, 1]+rad[i, j+1]*corner_mask[1, 1]
            kernel[row-1, 0] = temp
        elif i == row - 1 and j == col - 1:
            temp = rad[i-1, j-1]*corner_mask[0, 0]+rad[i, j-1]*corner_mask[1, 0]+rad[i-1, j]*corner_mask[0, 1]+rad[i, j]*corner_mask[1, 1]
            kernel[row-1, col-1] = temp
        elif i == 0 and 0 < j < col - 1:
            temp = rad[i, j-1]*row_mask[0, 0]+rad[i, j]*row_mask[0, 1]+rad[i, j+1]*row_mask[0, 2]+rad[i+1, j-1]*row_mask[1, 0]+rad[i+1, j]*row_mask[1, 1]+rad[i+1, j+1]*row_mask[1, 2]
            kernel[i, j] = temp
        elif i == row - 1 and 0 < j < col - 1:
            temp = rad[i-1, j-1]*row_mask[0, 0]+rad[i-1, j]*row_mask[0, 1]+rad[i-1, j+1]*row_mask[0, 2]+rad[i, j-1]*row_mask[1, 0]+rad[i, j]*row_mask[1, 1]+rad[i, j+1]*row_mask[1, 2]
            kernel[i, j] = temp
        elif 0 < i < row - 1 and j == 0:
            temp = rad[i-1, j]*col_mask[0, 0]+rad[i-1, j+1]*col_mask[0, 1]+rad[i, j]*col_mask[1, 0]+rad[i, j+1]*col_mask[1, 1]+rad[i+1, j]*col_mask[2, 0]+rad[i+1, j+1]*col_mask[2, 1]
            kernel[i, j] = temp
        elif 0 < i < row - 1 and j == col - 1:
            temp = rad[i-1, j-1]*col_mask[0, 0]+rad[i-1, j]*col_mask[0, 1]+rad[i, j-1]*col_mask[1, 0]+rad[i, j]*col_mask[1, 1]+rad[i+1, j-1]*col_mask[2, 0]+rad[i+1, j]*col_mask[2, 1]
            kernel[i, j] = temp
        else:
            temp = rad[i-1, j-1]*mask[0, 0]+rad[i-1, j]*mask[0, 1]+rad[i-1, j + 1]*mask[0, 2]+rad[i, j-1]*mask[1, 0]+ rad[i, j]*mask[1, 1]+rad[i, j + 1]*mask[1, 2]+rad[i + 1, j-1]*mask[2, 0]+rad[i + 1, j]*mask[2, 1]+rad[i + 1, j + 1]*mask[2, 2] 
            kernel[i, j]= temp 

kernel = kernel.astype(np.uint8) 
cv2.imwrite('mean_filer_rad.png', kernel)
```
